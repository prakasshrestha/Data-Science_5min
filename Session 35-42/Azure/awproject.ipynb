##Normalization

```python
from pyspark.sql.functions import col, min, max

data1 = [("shri", 27), ("pooja", 26), ("raj", 50)]
columns = ["Name", "Age"]

newdf = spark.createDataFrame(data1, columns)

age_min = newdf.select(min(col("Age"))).collect()[0][0]
age_max = newdf.select(max(col("Age"))).collect()[0][0]

df_normalized = newdf.withColumn("Age_Normalized", (col("Age") - age_min) / (age_max - age_min))

df_normalized.show()
```

##Standardization

```python
age_mean = newdf.select(mean(col("Age"))).collect()[0][0]
age_stddev = newdf.select(stddev(col("Age"))).collect()[0][0]

df_standardized = newdf.withColumn("Age_Standardized", (col("Age") - age_mean) / age_stddev)

df_standardized.show()
```

##Lable encoding

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="job_domain", outputCol="job_Index")

df_indexed = indexer.fit(dfemp).transform(dfemp)
df_indexed.display()
```

##One Hot Encoding

```python
from pyspark.ml.feature import OneHotEncoder

encoder = OneHotEncoder(inputCol="job_Index", outputCol="job_encode")

df_encoded = encoder.fit(df_indexed).transform(df_indexed)
df_encoded.display()    
```
